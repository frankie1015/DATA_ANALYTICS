```python
import pandas as pd

# Load the dataset with the specified encoding
file_path = "C:/Users/FREDDIE/Downloads/seoul+bike+sharing+demand/SeoulBikeData.csv"
seoul_bike_data = pd.read_csv(file_path, encoding='latin1')

# Print column names
print(seoul_bike_data.columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data Visualization
# Explore the relationship between bike rental counts and temperature
plt.figure(figsize=(10, 6))
sns.scatterplot(data=seoul_bike_data, x='Temperature(°C)', y='Rented Bike Count')
plt.title('Bike Rental Counts vs. Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Rented Bike Count')
plt.tight_layout()
plt.show()



```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data Visualization

# Plot time series of bike rental counts over the years
plt.figure(figsize=(12, 6))
sns.lineplot(data=seoul_bike_data, x='Date', y='Rented Bike Count')
plt.title('Bike Rental Counts Over Time')
plt.xlabel('Date')
plt.ylabel('Rented Bike Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Box plot to explore bike rental counts across different seasons
plt.figure(figsize=(10, 6))
sns.boxplot(data=seoul_bike_data, x='Seasons', y='Rented Bike Count')
plt.title('Bike Rental Counts Across Seasons')
plt.xlabel('Seasons')
plt.ylabel('Rented Bike Count')
plt.tight_layout()
plt.show()
```
